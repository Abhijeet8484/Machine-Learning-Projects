```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens)==2:
        return ((float(tokens[0]) + float(tokens[1])) / 2)
    try:
        return float(x)
    except:
        return None
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        st = np.std(subdf['price_per_sqft'])
        reduced_df = subdf[ (subdf['price_per_sqft'] > (m-st) ) & (subdf['price_per_sqft'] <= (m+st) ) ]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
        
    return df_out
```

```python
df7 = remove_pps_outliers(df6)

df7.shape
```

```python
# Finding the Houses where 2BHK price is higher than 3 BHK in a particular location
# Plotting a Scatter Plot

def plot_scatter(df, location):
    bhk2 = df[(df.location == location) & (df.bhk==2)]
    bhk3 = df[(df.location == location) & (df.bhk==3)]
    
    matplotlib.rcParams['figure.figsize'] = (10, 5)
    
    plt.scatter(bhk2.total_sqft, bhk2.price, color='red', label = '2 BHK', s = 50)
    plt.scatter(bhk3.total_sqft, bhk3.price, marker='x', label = '3 BHK', s = 50)
    
    plt.xlabel('Total Sq Ft Area')
    plt.ylabel('Price Per Sq Ft')
    
    plt.title(location)
    
    plt.legend()

plot_scatter(df7, 'Rajaji Nagar')    
```

```python
# Removing the 2 BHK outliers stated above

def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_sqft),
                'std' : np.std(bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0]
            }
            
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5:
                exclude_indices  = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
    
    return df.drop(exclude_indices, axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
# Removing the above Bathroom outliers

df9 = df8[df8.bath < (df8.bhk+2)]
df9.shape
```

```python
# Droppping cols : size(size & bhk col give same info), price_per_sqft(was only used for Outlier detection)

df10 = df9.drop(['size', 'price_per_sqft'], axis = 'columns')
df10.head()
```

```python
# One Hot Encoding on location col

dummies = pd.get_dummies(df10['location'])
dummies
```

```python
# Concating df11 with dummies cols but dropping the 'other' col in dummies bcoz in one hot encoding it can be considered as 0 0 0...

df11 = pd.concat([df10, dummies.drop('other', axis='columns')], axis='columns')
df11.head()
```

```python
# Dropping the location col as it's no longer needed due to one hot encoding

df12 = df11.drop('location', axis='columns')
df12.head()
```

```python
df12.shape
```

```python
# Dropping the dependent column i.e Price

X = df12.drop('price', axis='columns')
X.head()
```

```python
y = df12.price
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

lr_clf = LinearRegression()

lr_clf.fit(X_train, y_train)
#lr_clf.score(X_train, y_train)

lr_clf.score(X_test, y_test)
```

```python
##

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
# Finding Best Model using Grid SearchCV

from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridSearch(X, y):
    algos = {
        'linear_regression' : {
        'model' : LinearRegression(),
        'params' : {
            'normalize' : [True, False]
        }
        },
        
        'lasso' : {
        'model' : Lasso(),
        'params' : {
            'alpha' : [1,2],
            'selection' : ['random', 'cyclic']
        }
        },
        
        'decision_tree' : {
        'model' : DecisionTreeRegressor(),
        'params' : {
                'criterion' : ['mse', 'friedman_mse'],
                'splitter' : ['best', 'random']
            }
        }
    }
    

    scores = []

    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
      gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
      gs.fit(X, y)

      scores.append(
          {
              'model' : algo_name,
              'best_score' : gs.best_score_,
              'best_params' : gs.best_params_
          }
      )

    return pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'] )

find_best_model_using_gridSearch(X, y)
```

```python
# From above cell Linear Regx is the best model, so using in in below function

def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
    
```
